In this notebook, we'll investigate using unsupervised learning algorithms on MNIST digit data. First, let's import the wrapper for the data set:

```python
import digit_features as df
```

One interesting avenue of investigation is online unsupervised algorithms. That is, with a stream of data (in this case, handwritten digits) coming in, we can try to separate the data into clusters without having seen all of it. The online version of the popular K-means algorithm is one way of doing this - we fix a number of clusters, K, and set the values of the first K points to the means of the K clusters. Then, for each subsequent point, we assign it to the cluster for which the mean is closest to the point's value. When this happens, we modify the cluster mean in a weighted manner - if there have been M points previously assigned the cluster, the incoming point gets weight one, and the previous mean gets weight M, and the new mean is just the weighted average. Let's try this approach on the MNIST data with various numbers of clusters fixed:

```python
from SeqKMeans import SeqKMeans
```

Here, for every number of clusters, we can see the changing means of the clusters as every 10% of the data is processed.

```python
skm = SeqKMeans(df.flatPixelTrainData())
```

```python
skm.trainMeans(2)
skm.trainMeans(3)
skm.trainMeans(5)
skm.trainMeans(10)
skm.trainMeans(15)
skm.trainMeans(20)
```

Especially for smaller numbers of clusters, we can see that there tend to be 1-3 clusters with very large numbers of points assigned to them - evidently, they become a sort of fuzzy average, which is often closer to an incoming data point than even members of that point's class. Equally problematic is the fact that unusually shaped digits can be so dissimilar to other points that if a cluster is initialized with  one of them as the mean, it may end up with no additional points being assigned to it, since the "fuzzy average" clusters consistently score closer to incoming points on distance. (Here, we are using the Euclidian distance metric - see SeqKMeans.py for details)
In short, this method does succeed in separating out digits into clusters to some degree, but by no means is it close to good enough for accurate classification. We can also assume that there are ~5 types of each digit and try separating the images into 50 classes:

```python
skm.trainMeans(50)
```

It looks like running sequential K-means on 50 clusters produces some sensible clusters but also a significant amount of singletons and clusters with <5 points.

We can also try an approach where we start with a single cluster, and for incoming points, if the point is within a distance threshold of the cluster, assign it to the cluster, and otherwise form a new cluster with that point. Let's try this with different thresholds:

```python
#Thresholds become more permissive as number gets higher, meaning fewer new clusters are generated
skm.trainMeansGrowing(10.0)
skm.trainMeansGrowing(11.0)
skm.trainMeansGrowing(12.0)
```

Evidently, this does not solve the problem of clusters ending up with either too many or too few points, regardless of the threshold value.

Now, let's try using a different metric for classification and distance - instead of computing Euclidian distances between the cluster-average vectors and the data points, we instead model the cluster means as vectors of Bernoulli probabilities, and use the Bernoulli likelihood of a binary feature vector relative to the cluster probability vector as the fit metric. First, we use this metric to try training a fixed number of clusters as we did with the Euclidian distance. Here, epsilon is the "default" Bernoulli probability when the actual value would be 0, as with a 1 feature with 0 probability, or 0 feature with 1 probability.

EDIT: now we are not using epsilon, we are adding 2 both to the number of 1s and number of 0s.

```python
skm.trainMeans(5, useBernoulli=True, epsilon=.01)
skm.trainMeans(10, useBernoulli=True, epsilon=.01)
skm.trainMeans(20, useBernoulli=True, epsilon=.01)

skm.trainMeans(5, useBernoulli=True, epsilon=.1)
skm.trainMeans(10, useBernoulli=True, epsilon=.1)
skm.trainMeans(20, useBernoulli=True, epsilon=.1)

skm.trainMeans(5, useBernoulli=True, epsilon=.2)
skm.trainMeans(10, useBernoulli=True, epsilon=.2)
skm.trainMeans(20, useBernoulli=True, epsilon=.2)
```

To finish the comparison, we can try grouping the digits into 50 clusters as before, but with the Bernoulli metric:

```python
skm.trainMeans(50, useBernoulli=True, epsilon=.2)
```

Finally, we try to cluster the digits using growing K-means with the Bernoulli metric. That is, for each cluster, we keep track of the likelihoods for every point in the cluster relative to its mean. For each new digit, we find the best cluster, and if the new digit is within 3 standard deviations of the mean likelihood, add it to the cluster, otherwise start a new cluster.

```python
skm.trainMeansGrowingBernoulli(std_cutoff=3)
skm.trainMeansGrowingBernoulli(std_cutoff=3)
skm.trainMeansGrowingBernoulli(std_cutoff=3)
```

```python
skm.trainMeansGrowingBernoulli(std_cutoff=2.5)
skm.trainMeansGrowingBernoulli(std_cutoff=2.5)
skm.trainMeansGrowingBernoulli(std_cutoff=2.5)
```

Evidently, this attempt to dynamically determine clusters is not ideal, but does generate somewhat reasonable clusters
