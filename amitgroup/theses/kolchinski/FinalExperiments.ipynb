```python
vecLen = 100
highProb = 0.5
bgdProb = 0

trainData = sf.synthData(ptsPerClass=100,N=vecLen,M=vecLen/2,highOnly=True,highP=highProb,bgdP=bgdProb)

synthPegasos = PegasosSVM(2)
#Third argument is number of iterations
weightHistory = synthPegasos.trainOnSet(df.labeled(trainData, 2), trainData, 2000)

testData = df.labeled(sf.synthData(ptsPerClass=100,N=vecLen,M=vecLen/2,highOnly=True,highP=highProb,bgdP=bgdProb), 2)
synthPegasos.testOnSet(testData)

#print weightHistory.shape
avgHighWtsPosClass = weightHistory[1][:,:vecLen/2].mean(axis=1)
avgBgdWtsPosClass = weightHistory[1][:,vecLen/2:].mean(axis=1)
avgHighWtsNegClass = -1.0*weightHistory[0][:,:vecLen/2].mean(axis=1)
avgBgdWtsNegClass = -1.0*weightHistory[0][:,vecLen/2:].mean(axis=1)

Xs = np.arange(len(weightHistory[1]))*10

plt.figure(figsize=(12,10))
plt.title("Convergence of SVM weights")

plt.plot(Xs,avgHighWtsPosClass,'g-',label='Avg weight over high-p region, pos class')
plt.plot(Xs,avgBgdWtsPosClass,'b-',label='Avg weight over bgd-p region, pos class')
plt.plot(Xs,avgHighWtsNegClass,'g--',label='-1 * Avg weight over high-p region, neg class')
plt.plot(Xs,avgBgdWtsNegClass,'b--',label='-1 * Avg weight over bgd-p region, neg class')
plt.legend(loc=0)
plt.show()
```
